```python
from util import *

g = gen_weighted_undirected_graph_fsq(lambda e, nodes: float(1)/compute_distance(e, nodes))

import matplotlib.pyplot as plt
%matplotlib inline

import networkx as nx
import numpy as np
from math import ceil
```

```python
g.node[0]
```

```python
g.degree(0)
```

```python
g.degree(0, weight="weight")
```

```python
degrees = np.bincount(map(lambda x: ceil(x), g.degree(weight="weight").itervalues()))
```

```python
plot(degrees, "weighted_degree", "Weighted Degree")
```

```python
total_checkins = [g.node[n]["total_checkins"] for n in g.nodes()]
```

```python
degrees_in_order = [g.degree(n, weight="weight") for n in g.nodes()]
```

```python
plt.plot(degrees_in_order, total_checkins, 'ro')
```

```python
unweighted_degrees_in_order = [g.degree(n) for n in g.nodes()]
```

```python
plt.plot(unweighted_degrees_in_order, total_checkins, 'ro')
```

```python
small_checkins = [n for n in g.nodes() if g.node[n]["total_checkins"] < 5000]
```

```python
len(small_checkins)
```

```python
uw_d_sm = [g.degree(n) for n in small_checkins]
uw_c_sm = [g.node[n]["total_checkins"] for n in small_checkins]
```

```python
plt.plot(uw_d_sm, uw_c_sm, 'ro')
```

```python
w_d_sm = [g.degree(n, weight="weight") for n in small_checkins]
plt.plot(w_d_sm, uw_c_sm, 'ro')
```

```python
uw_degrees = np.bincount(map(float, g.degree()))
plot(uw_degrees, "unweighted_degree", "Unweighted Degree")
```

```python
s = float(uw_degrees.sum())
cdf = uw_degrees.cumsum(0)/s
ccdf = 1-cdf
plt.plot(range(len(ccdf)),ccdf,'bo')
plt.ylim([0,1])
plt.ylabel('CCDF')
plt.xlabel("unweighted degree")
```

```python
s = float(degrees.sum())
cdf = degrees.cumsum(0)/s
ccdf = 1-cdf
plt.plot(range(len(ccdf)),ccdf,'bo')
plt.ylim([0,1])
plt.ylabel('CCDF')
plt.xlabel("weighted degree")
```
