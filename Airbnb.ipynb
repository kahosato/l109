```python
AIRBNB_DATA_DIR = "/Users/kaho/projects/data_l109/airbnb/"
import os
import csv
```

```python
listings = {}
keys = ["neighbourhood_group", "neighbourhood", "latitude", "longitude"]
floats = set(["latitude", "longitude"])
with open(os.path.join(AIRBNB_DATA_DIR, "listings_simple.csv"), 'r') as f:
    r = csv.DictReader(f)
    for row in r:
        info = {}
        for k in keys:
            if k in floats:
                info[k] = float(row[k])
            else:
                info[k] = row[k]
        listings[row["id"]] = info
```

```python
fsq_partions = read_partitions()
fsq_nodes = gen_fsq_nodes()
listings = gen_listings(keys, floats)
def distance(from_node, to_node):
    return pow(pow((from_node["longitude"] - to_node["longtitude"]), 2)
               + pow((from_node["latitude"] - to_node["latitude"]), 2), 0.5)
with open("airbnb_community.txt", 'w') as f:
    for l_id, listing in listings.iteritems():
        first_id, first_node = fsq_nodes.iteritems().next()
        min_distance = distance(listing, first_node)
        neighbour = first_id
        for nid, fsq_node in fsq_nodes.iteritems():
            d = distance(listing, fsq_node)
            if min_distance > d:
                min_distance = d
                neighbour = nid
        f.write("{} {} {} {}\n".format(l_id, fsq_partions[neighbour], neighbour, min_distance))
```
